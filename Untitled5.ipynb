```python

import multiprocessing
import time
import random
from datetime import datetime

def worker():
    time.sleep(random.random())
    q.put(f"{multiprocessing.current_process().name}: {datetime.now().strftime('%H:%M:%S.%F')}")

if __name__ == "__main__":
    q = multiprocessing.Queue()
    processes = []
    for _ in range(3):
        p = multiprocessing.Process(target = worker, args=(q, ))
        processes.append(p)
        p.start()

    for p in processes:
        p.join()
    while not q.empty():
        print(q.get())
    print("done")
```
