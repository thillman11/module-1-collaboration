```python
import sqlite3, csv

conn = sqlite3.connect('books.db')
cur = conn.cursor()

cur.execute("""
create table books (
    title TEXT,
    author TEXT,
    year INTEGER
    )
""")
with open('books2.csv', newline='', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    rows = [(r['title'], r['author'], int(r['year'])) for r in reader]

cur.execute(" delete from books")
cur.executemany(" insert into books values (?, ?, ?)", rows)
conn.commit()

for row in cur.execute("select * from books order by year"):
    print(row)
```
