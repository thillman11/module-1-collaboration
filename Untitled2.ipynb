```python
secret = 5
guess = 2

if guess < secret:
    print('too low')
elif guess > secret:
    print('too high')
else:
    print('just right')
```

```python
small = True
green = True

if small:
    if green:
        print('pea')
    else:
        print('cherry')
else: 
    if green:
        print('watermelon')
    else: 
        print('pumpkin')
```

```python
guess_me = 7
number = 1

while True:
    if number < guess_me:
        print('too low')
    elif number == guess_me:
        print ('found it!')
        break
    else:
        print('oops')

    number+= 1
```

```python
guess_me = 5

for number in range(10):
    if number < guess_me:
        print('too low')
    elif number == guess_me:
        print('found it!')
        break
    else:
        print('oops')
        break
```
